```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.head(1)
```

```python
movies.info()
# genres
# keywords 
# overview
# title
# cast
# crew
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['cast'][0]
```

```python
import ast
def convert_3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_3)
```

```python
movies.head(1)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies['tags']
```

```python
df = movies[['movie_id', 'title', 'tags']]
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x: x.lower())
df['tags'][0]
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector
```

```python
cv.get_feature_names_out()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# For to remove same meaning words to their root words.
def stem(text):
    L = []
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df.head()
```

```python
cv.get_feature_names_out()
```

```python
# high dimensional data is inversely proportional to Ecludeian distance (means the disatance betwwen two points not there angle).
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity[1]
```

```python
similarity[0]
```

```python
# Finding Index of movie given in recommend function then after we check its similarities
movie_index = df[df['title'] == 'Batman Begins'].index[0]
print(movie_index)
```

```python
# For making a list of top 5 similar movie with maintining focus on index by enumerate class/function and then sorted in reverse and maintaining focus
# to start from x[1] so that sorting will done from similarities 
movie_list = sorted(list(enumerate(similarity[0])),reverse=True, key = lambda x:x[1])[1:6]
print(movie_list)
```

```python
def recommend(movie):
    movie_index = df[df['title'] ==  movie].index[0]
    # print(movie_index)
    distances = similarity[movie_index]
    # print(distances)
    movie_list = sorted(list(enumerate(distances)),reverse=True, key = lambda x:x[1])
    # print(movie_list)
    
    for i in movie_list[1:6]:
        print(df.iloc[i[0]].title)
    
```
